```python
import numpy as np, os
from tqdm import tqdm
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input
from tensorflow.keras.preprocessing import image

# Load ResNet50 model (pretrained on ImageNet)
model = ResNet50(weights='imagenet', include_top=False, pooling='avg')

def extract_features(img_path, model):
    img = image.load_img(img_path, target_size=(224, 224))
    img_data = image.img_to_array(img)
    img_data = np.expand_dims(img_data, axis=0)
    img_data = preprocess_input(img_data)
    features = model.predict(img_data, verbose=0)
    return features.flatten()

# Gather all image paths
image_paths = []
for root, _, files_in_dir in os.walk("fashion_dataset"):
    for fname in files_in_dir:
        if fname.lower().endswith(('.jpg', '.jpeg', '.png')):
            image_paths.append(os.path.join(root, fname))

print(f"üñº Total images ready: {len(image_paths)}")

# Extract and store features
features_list = []
for img_path in tqdm(image_paths):
    try:
        features = extract_features(img_path, model)
        features_list.append(features)
    except Exception as e:
        print(f"‚ö†Ô∏è Skipped {img_path}: {e}")

features_list = np.array(features_list)
print("‚úÖ Feature extraction completed.")
```

```python
np.save("features.npy", features_list)
np.save("image_paths.npy", np.array(image_paths))
print("‚úÖ Features saved for reuse.")
```

```python
features_list = np.load("features.npy")
image_paths = np.load("image_paths.npy", allow_pickle=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_similar_images(uploaded_img_path, features_list, image_paths, model, top_n=5):
    uploaded_features = extract_features(uploaded_img_path, model)
    similarities = cosine_similarity([uploaded_features], features_list)[0]
    indices = np.argsort(similarities)[::-1][1:top_n+1]
    return [image_paths[i] for i in indices]
```

```python
from google.colab import files
from PIL import Image
from IPython.display import display
import matplotlib.pyplot as plt

# Upload a test image
uploaded = files.upload()
uploaded_img_path = list(uploaded.keys())[0]

# Get recommendations
recommended_images = recommend_similar_images(
    uploaded_img_path, features_list, image_paths, model, top_n=5
)

# Display uploaded image
print("üñº Uploaded Image:")
display(Image.open(uploaded_img_path).resize((224, 224)))

# Display 5 recommended images side-by-side
print("\nüëó Recommended Similar Images:")
plt.figure(figsize=(15, 5))
for i, path in enumerate(recommended_images):
    plt.subplot(1, 5, i+1)
    img = Image.open(path)
    plt.imshow(img)
    plt.axis('off')
plt.show()
```
